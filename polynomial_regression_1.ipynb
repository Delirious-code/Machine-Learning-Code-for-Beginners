```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/Users/apple/Desktop/NLP Run/Udemy Machine Learning/Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv")
#Connect it to the Position Salaries CSV
```

```python
x = dataset.iloc[ : , 1:-1 ].values
y = dataset.iloc[ : , -1 ].values
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

# Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=5)
x_poly = pf.fit_transform(x)
lr_2 = LinearRegression()
lr_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,lr.predict(x),color = 'blue')
plt.title("Truth or Bluff (Linear Regression Model)")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,lr_2.predict(x_poly),color = 'blue')
plt.title("Truth or Bluff (Polynomial Regression Model)")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Visualising it better

```python
x_grid = np.arange(min(x), max(x),0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.plot(x_grid,lr_2.predict(pf.fit_transform(x_grid)),color = 'blue')
plt.scatter(x,y,color = "red")
plt.title("Truth or Bluff (Polynomial Regression Model, better visualized)")
plt.xlabel('Level')
plt.ylabel('Salary')
```

# Predicting a New Result with Linear Regression

```python
lr.predict([[6.5]])
```

# Predicting the New Result with Polynomial Regression

```python
lr_2.predict(pf.fit_transform([[6.5]]))
```
