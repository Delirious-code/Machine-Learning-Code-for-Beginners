```python
import pandas as pd

data = pd.read_csv("/Users/apple/Desktop/NLP Run/Datasets - Intel/surat_uncleaned.csv")

data.head()
```

```python
pd.set_option('display.max_rows', None)
column_name = 'facing'
missing_values = data['facing'].isnull()
print(data[missing_values])
```

```python
data_clean = data.dropna(subset=['facing'])

data.head()
```

```python
def clean_facing_directions(value):
    if isinstance(value,str):
        value = value.strip()
        value = value.replace(" -","-").replace("- ","-")
        value = value.lower()
        value = value.replace("south-east","southwest")
        value = value.replace("south-west","southeast")
        value = value.replace("north-east","northeast")
        value = value.replace("north-west","northwest")
        return value
    else:
        return value
    
    
data['facing'] = data['facing'].apply(clean_facing_directions)

data.head()
                
```

```python
def categorizing_facing(value):
    directions = ['north', 'south', 'east', 'west', 'northeast', 'northwest', 'southeast', 'southwest']
    
    if isinstance(value,str):
        if value in directions:
            return value
        elif 'park' in value or 'garden/park' in value or 'main road' in value:
            return "near landmark"
        else:
            return 'other'
    elif pd.isnull(value):
        return "Incomplete"
    else:
        return "Invalid"

data['facing'] = data['facing'].apply(categorizing_facing)

data.head()            
```

```python
dropped_data = ['property_name','status','description']
data.drop(dropped_data,axis=1)
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data_grouped = data.groupby(['areaWithType', 'facing']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))  # Increase the figure size for better visibility
sns.barplot(x='areaWithType', y='count', hue='facing', data=data_grouped)
plt.xlabel('Area Type')
plt.ylabel('Count of Facing Directions')
plt.title('Count of Facing Directions by Area Type')
plt.xticks(rotation=45)
plt.show()
```
