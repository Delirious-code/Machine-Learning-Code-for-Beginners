```python

gen_map = {'Female':0, 'Male' : 1,'Prefer not to say' : 2}
data['What is your gender?'] = data['What is your gender?'].map(gen_map)

year_map = {'Freshman (1st year)' : 1, 'Sophomore (2nd year)' : 2, 'Junior (3rd year)' : 3, 'Senior (4th year or higher)' : 4}
data['What year of study are you currently in?'] = data['What year of study are you currently in?'].map(year_map)

age_map = {'18-20' : 19,'21-23' : 22, '24-26' : 25, '27 and above' : 27}
data['What is your age?'] = data['What is your age?'].map(age_map)
major_mapping = {
    'School of Sciences': 1,
    'School of Arts and Humanities': 2,
    'School of Business and Management': 3,
    'School of Psychological Sciences': 4,
    'School of Law': 5,
    'School of Engineering and Technology': 6,
    'School of Commerce, Finance and Accountancy': 7
}
data['What is your major or field of study?'] = data['What is your major or field of study?'].map(major_mapping)

# day_map = {'Less than 4 hours' : 3-4,'More than 7 hours' : 7-8}
# data['On average, how many hours do you sleep on a typical weeknight?'] = data['On average, how many hours do you sleep on a typical weeknight?'].map(day_map)
data = data.dropna(subset=['What is your major or field of study?'])
data = data.dropna(subset=['What is your current GPA?'])

data.head(100)
```

```python


def average_range(value):
    value = value.lower().replace('hours', '').strip()
      
    if value == "less than 4":
        return (3 + 4) / 2  # Average of 3 and 4
    elif value == "more than 7":
        return (7 + 8) / 2  # Average of 7 and 8
    elif isinstance(value, str) and '-' in value:
        # Split the range (e.g., '4-5') into two integers
        low, high = map(int, value.split('-'))
        return (low + high) / 2  # Return the average of the two values
    return value  # Return unchanged if no condition matches

# Apply the function to the relevant column in the DataFrame
data['On average, how many hours do you sleep on a typical weeknight?'] = data['On average, how many hours do you sleep on a typical weeknight?'].apply(average_range)
data['On average, how many hours do you sleep on weekends (Saturday and Sunday combined)'] = data['On average, how many hours do you sleep on weekends (Saturday and Sunday combined)'].apply(average_range)
data.head()
```

```python
def avg_gpa(val):
    val = val.lower().strip()
    if isinstance(val, str) and '-' in val:
        # Split the range (e.g., '4-5') into two integers
        low, high = map(float, val.split('-'))
        return (low + high) / 2  # Return the average of the two values
    return val # Return unchanged if no condition matches
    
data['What is your current GPA?'] = data['What is your current GPA?'].apply(avg_gpa)
data['How much a person sleeps in the entire week on an average'] = data['On average, how many hours do you sleep on a typical weeknight?'] + data['On average, how many hours do you sleep on weekends (Saturday and Sunday combined)']
data.head()
```

```python
x = data['What is your current GPA?']
y = data['How much a person sleeps in the entire week on an average']
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
if isinstance(xtrain, pd.Series):
    xtrain = xtrain.to_frame()
    
if isinstance(xtest, pd.Series):
    xtest = xtest.to_frame()
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn_regressor = KNeighborsRegressor(n_neighbors=3)  # You can adjust the number of neighbors (n_neighbors)
knn_regressor.fit(xtrain, ytrain)
```

```python
y_pred = knn_regressor.predict(xtest)
```
